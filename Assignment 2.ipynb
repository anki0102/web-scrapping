```python
# accessing the website
# defining the input in the search field

driver.get('https://www.flipkart.com/')
user_inp= input('enter the product you want to search:')
```

```python
# accessing the website

driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb-includes- earpods-power- adapter/p/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKC TSVZAXUHGREPBFGI&marketplace')
```

```python
#clicjing the All reviews buttons through .click()

search_button= driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div/div/div[5]/div/a/div/span')
search_button.click()
```

# Program 10

Scraping first 10 laptops data from Amazon.in

Using Filters as “Intel Core i7” and “Intel Core i9”

```python
#importing the libraries
import selenium
import pandas as pd
import time
from selenium import webdriver
```

```python
#connecting to chromedriver

driver = webdriver.Chrome(r'/Users/ankitraj1/chromedriver')
```

```python
# accessing the website
# defining the input in the search field

driver.get('https://www.amazon.in/')
user_inp= input('enter the product you want to search:')
```

```python
# giving the input in the search field by send_keys
#clicking the search button through .click()

search_bar = driver.find_element_by_id("twotabsearchtextbox")
search_bar.clear()
search_bar.send_keys(user_inp)
time.sleep(5)
search_button= driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]/span/input')
search_button.click()
                                   
```

```python
#selecting the filter button "intel core i7" through .click()

search_button= driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/div/label/i')
search_button.click()
  
```

```python
# declaring the desired output

title=[]
ratings=[]
price=[]
```

```python
#accessing the items that are required

for j in driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]'):
        title.append(j.text)
for k in driver.find_elements_by_xpath('//div[@class="a-row a-size-small"]/span[2]/a/span'):
        ratings.append(k.text)
for l in driver.find_elements_by_xpath('//span[@class="a-price-whole"]'):
        price.append(l.text)
```

```python
#creating the data frame for the outputs

data = list(zip(title,ratings,price))
df = pd.DataFrame (data,columns=['title','ratings','price'])
```

```python
#displaying the dataframe

df.head(10)
```

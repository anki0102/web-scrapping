# Program - 2

Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
from selenium.common.exceptions import NoSuchElementException, ElementNotVisibleException,StaleElementReferenceException
```

```python
# Getting href of all pages 
URL=[]
urls=driver.find_elements_by_xpath("//div[@class='a-section a-spacing-none a-padding-base']/div[1]/ul[1]/li/a")
for i in urls:
    URL.append(i.get_attribute('href'))
```

```python
URL[0:3]
```

```python
# We extract href of all the laptops in first three pages and store in list URL2
URL2=[]
for j in URL[0:3]:
    driver.get(j)
    url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in url:
        URL2.append(i.get_attribute('href'))
```

```python
#Create empty list to store scrapped data.
NAME=[]
PRICE=[]
DELI=[]
STOCK=[]
REV=[]
RATING=[]
```

```python
#We extract desired data in text form and store it in respective lists.
for j in URL2:
    driver.get(j)
    try:
        name=driver.find_element_by_id("productTitle")
        NAME.append(name.text)
    except NoSuchElementException:
        NAME.append('--')
    try:
        price=driver.find_element_by_xpath("//td[@class='a-span12']/span[1]")
        PRICE.append(price.text)  
    except NoSuchElementException :
        PRICE.append('--') 
    try:
        deli=driver.find_element_by_id('ddmDeliveryMessage')
        DELI.append(deli.text)
    except NoSuchElementException : 
        DELI.append('--')
    try:
        stock=driver.find_element_by_id("availability")
        STOCK.append(stock.text)
    except NoSuchElementException :
        STOCK.append('--')
    try:
        rev=driver.find_element_by_id('acrCustomerReviewText')
        REV.append(rev.text)
    except NoSuchElementException:
        REV.append('--')
    try:
        driver.get(j)
        cl=driver.find_element_by_id('acrCustomerReviewText').click()
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        RATING.append(rating.text)
    except NoSuchElementException:
        RATING.append("--")
```

```python
print(len(URL2),len(PRICE),len(DELI),len(STOCK),len(REV),len(RATING))
```

```python
import pandas as pd
Laptop=pd.DataFrame()
```

```python
Laptop['NAME']=NAME[:]
Laptop['Price']=PRICE[:]
Laptop['Delivery Time']=DELI[:]
Laptop['Stock_Avaibility']=STOCK[:]
Laptop['Rating']=RATING[:]
Laptop['No._of_Reviews']=REV[:]
Laptop['Product_URL']=URL2[:]
```
